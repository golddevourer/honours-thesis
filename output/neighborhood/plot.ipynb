```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams["figure.figsize"] = (10,10)
plt.subplot(2,2,1)
n1 = np.loadtxt('n_cells_birth_0.1_all.txt')
n2 = np.loadtxt('n_cells_birth_0.1_first.txt')
n3 = np.loadtxt('n_cells_birth_0.1_second.txt')
y1=n1[0,:]/np.sum(n1,0)
y2=n2[0,:]/np.sum(n2,0)
y3=n3[0,:]/np.sum(n3,0)
plt.plot(y1,label="global")
plt.plot(y2,label="first neighbor")
plt.plot(y3,label="second neighbor")
plt.ylabel("tumor size/%")
plt.xlabel("time/mcs")
plt.legend()
plt.title("(a)")

plt.subplot(2,2,2)
n1 = np.loadtxt('n_cells_death_0.003_0.01_all.txt')
n2 = np.loadtxt('n_cells_death_0.003_0.01_first.txt')
n3 = np.loadtxt('n_cells_death_0.003_0.01_second.txt')
y1=n1[0,:]/np.sum(n1,0)
y2=n2[0,:]/np.sum(n2,0)
y3=n3[0,:]/np.sum(n3,0)
plt.plot(y1,label="global")
plt.plot(y2,label="local")
plt.plot(y3,label="second neighbor")
plt.ylabel("tumor size/%")
plt.xlabel("time/mcs")
plt.legend()
plt.title("(b)")

plt.subplot(2,2,3)
n1 = np.loadtxt('n_cells_stiffness_0.6_0.2_all.txt')
n2 = np.loadtxt('n_cells_stiffness_0.6_0.2_first.txt')
y1=n1[0,:]/np.sum(n1,0)
y2=n2[0,:]/np.sum(n2,0)
plt.plot(y1,label="global")
plt.plot(y2,label="local")
plt.ylabel("tumor size/%")
plt.xlabel("time/mcs")
plt.legend()
plt.title("(c)")

plt.subplot(2,2,4)
n1 = np.loadtxt('n_cells_contractility_all.txt')
n2 = np.loadtxt('n_cells_contractility_first.txt')
y1=n1[0,:]/np.sum(n1,0)
y2=n2[0,:]/np.sum(n2,0)
plt.plot(y1,label="global")
plt.plot(y2,label="first neighbor")
plt.ylabel("tumor size/%")
plt.xlabel("time/mcs")
plt.legend()
plt.title("(d)")
plt.show()
```
