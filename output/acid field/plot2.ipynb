```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams["figure.figsize"] = (7,7)
n1 = (np.loadtxt('n_cells_birth_0.1_1.txt')+np.loadtxt('n_cells_birth_1.txt'))/2
n2 = np.loadtxt('n_cells_birth_0.8.txt')
n3 = np.loadtxt('n_cells_birth_0.1_0.6.txt')
n4 = np.loadtxt('n_cells_birth_0.4.txt')
n5 = np.loadtxt('n_cells_birth_0.1_0.2.txt')
n6 = np.loadtxt('n_cells_birth_0.1_0.1.txt')
n7 = np.loadtxt('n_cells_birth_0.1_0.01.txt')
n8 = np.loadtxt('n_cells_birth_0.001.txt')
n9 = np.loadtxt('n_cells_birth_0.1_0.txt')
y1=n1[0,:]/np.sum(n1,0)
y2=n2[0,:]/np.sum(n2,0)
y3=n3[0,:]/np.sum(n3,0)
y4=n4[0,:]/np.sum(n4,0)
y5=n5[0,:]/np.sum(n5,0)
y6=n6[0,:]/np.sum(n6,0)
y7=n7[0,:]/np.sum(n7,0)
y8=n8[0,:]/np.sum(n8,0)
y9=n9[0,:]/np.sum(n9,0)
plt.plot(y1,label="diffusion=1")
plt.plot(y2,label="diffusion=0.8")
plt.plot(y3,label="diffusion=0.6")
plt.plot(y4,label="diffusion=0.4")
plt.plot(y5,label="diffusion=0.2")
plt.plot(y6,label="diffusion=0.1")
plt.plot(y7,label="diffusion=0.01")
plt.plot(y8,label="diffusion=0.001")
plt.plot(y9,label="diffusion=0")
plt.ylabel("tumor size/%")
plt.xlabel("time/mcs")
plt.legend()
plt.title("(a)birth rate")
plt.savefig("birth1.png") 
plt.show()
```

```python
x=np.arange(20000)
m1,b1=np.polyfit(x,y1,1)
m2,b2=np.polyfit(x,y2,1)
m3,b3=np.polyfit(x,y3,1)
m4,b4=np.polyfit(x,y4,1)
m5,b5=np.polyfit(x,y5,1)
m6,b6=np.polyfit(x,y6,1)
m7,b7=np.polyfit(x,y7,1)
m8,b8=np.polyfit(x,y8,1)
m9,b9=np.polyfit(x,y9,1)
plt.plot([1,0.8,0.6,0.4,0.2,0.1,0.01,0.001,0],[m1,m2,m3,m4,m5,m6,m7,m8,m9])
plt.scatter([1,0.8,0.6,0.4,0.2,0.1,0.01,0.001,0],[m1,m2,m3,m4,m5,m6,m7,m8,m9])

plt.ylabel("invasion velocity/%*mcs^-1")
plt.xlabel("diffusion constant")
plt.title("(a)birth rate")
plt.savefig("birth2.png") 
plt.show()
```
